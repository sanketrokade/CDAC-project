```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("./data/total_execution_time.csv")
```

```python
df1.head(15)
```

```python
plt.plot(df1['ncsf'],df1['seq'],'o-',label='Sequential')
plt.plot(df1['ncsf'],df1['MPI'],'o-',label='MPI')
plt.plot(df1['ncsf'],df1['CUDA+MPI'],'o-',label='CUDA + MPI')
plt.plot(df1['ncsf'],df1['CUDA+MPI+openMP'],'o-',label='CUDA + MPI + openMP')
plt.xlabel('Matrix size')
plt.ylabel('Total Execution time (sec)')
plt.legend()
```
